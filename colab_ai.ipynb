```python
# prompt: 마케팅 데이터 분석을 위한 임의의 데이터를 생성하되  행이 10000개 이상인 데이터를 만들 것, 데이터를 출력할 때 데이터프레임은 print가 아닌 display 구문으로 작성할 것
# 컬럼은 다음의 값을 사용하되 광고주 항목을 추가하고 광고주를 한국어로 임의로 10개 생성해서 빈도수가 정규분포 형태가 되도록 만들것
# columns = ['광고_지출', '노출_수', '클릭_수', '전환_수', '매출', '구매_단가', '고객_연령', '고객_성별', '제품_카테고리', '광고_플랫폼']

import pandas as pd
import numpy as np
from IPython.display import display

# 광고주 목록 생성
advertisers = ['삼성전자', 'LG전자', '현대자동차', '네이버', '카카오', '쿠팡', '배달의민족', 'GS25', 'CU', '이마트']

# 데이터 생성
n = 10000
data = {
    '광고_지출': np.random.randint(1000, 100000, n),
    '노출_수': np.random.randint(10000, 1000000, n),
    '클릭_수': np.random.randint(100, 10000, n),
    '전환_수': np.random.randint(10, 1000, n),
    '매출': np.random.randint(100000, 10000000, n),
    '구매_단가': np.random.randint(1000, 100000, n),
    '고객_연령': np.random.choice(range(20, 60), n),
    '고객_성별': np.random.choice(['남', '여'], n),
    '제품_카테고리': np.random.choice(['가전제품', '의류', '식품', '화장품', '여행'], n),
    '광고_플랫폼': np.random.choice(['네이버', '카카오', '구글', '페이스북', '인스타그램'], n),
}

# 광고주 컬럼 추가 (정규분포 형태)
advertiser_counts = np.random.normal(loc=len(advertisers)/2, scale=2, size=n).astype(int)
advertiser_counts = np.clip(advertiser_counts, 0, len(advertisers)-1)
data['광고주'] = [advertisers[i] for i in advertiser_counts]

# 데이터프레임 생성
df = pd.DataFrame(data)

# 데이터 출력
display(df)
```

```python
# 광고 플랫폼, 제품 카테고리, 고객 연령대별 클릭률, 전환율 계산
result = df.groupby(['광고_플랫폼', '제품_카테고리', pd.cut(df['고객_연령'], bins=[20, 30, 40, 50, 60])]) \
    .agg({'클릭_수': 'sum', '노출_수': 'sum', '전환_수': 'sum'})
result['클릭률'] = result['클릭_수'] / result['노출_수']
result['전환율'] = result['전환_수'] / result['클릭_수']
display(result)

# ANOVA 분석 (예시: 광고 플랫폼에 따른 클릭률 차이 검정)
from scipy import stats
group1 = df[df['광고_플랫폼'] == '네이버']['클릭_수'] / df[df['광고_플랫폼'] == '네이버']['노출_수']
group2 = df[df['광고_플랫폼'] == '카카오']['클릭_수'] / df[df['광고_플랫폼'] == '카카오']['노출_수']
# (다른 광고 플랫폼에 대한 그룹 생성)
group3 = df[df['광고_플랫폼'] == '페이스북']['클릭_수'] / df[df['광고_플랫폼'] == '페이스북']['노출_수']  # 예시: 페이스북
group4 = df[df['광고_플랫폼'] == '인스타그램']['클릭_수'] / df[df['광고_플랫폼'] == '인스타그램']['노출_수']  # 예시: 인스타그램

# '...'를 비교하려는 실제 그룹으로 바꿉니다.
f_statistic, p_value = stats.f_oneway(group1, group2, group3, group4)
print("F-statistic: {:.2f}, p-value: {:.3f}".format(f_statistic, p_value))
```
